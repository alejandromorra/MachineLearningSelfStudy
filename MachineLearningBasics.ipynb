```python
#Data Cleaning
```

```python
teams[teams.isnull().any(axis=1)] #Finding Zeros
```

```python
teams= teams.dropna() #drops any na values
```

```python
teams[teams.isnull().any(axis=1)] #Confirmed all na values are dropped
```

```python
#Split data
```

```python
# Split data must respect the time series asspect of data
```

```python
train = teams[teams["year"]<2012].copy()
test = teams[teams["year"] >= 2012].copy()
```

```python
train.shape
```

```python
test.shape
```

```python
splitpercent = test.shape[0]/train.shape[0]
splitpercent #About 75% to 25% 
```

```python
from sklearn.linear_model import LinearRegression
```

```python
reg = LinearRegression()
```

```python
predictors = ["athletes", "prev_medals"]
target = "medals"
```

```python
reg.fit(train[predictors], train["medals"])
```

```python
predictions = reg.predict(test[predictors])
```

```python
test["predictions"]=predictions #amend to test dataset
```

```python
test #note negative or decimal medals does not make sense in context
```

```python
test.loc[test["predictions"]< 0, "predictions"] = 0 #replace neg predictions with 0
```

```python
test["predictions"] = test["predictions"].round()
```

```python
from sklearn.metrics import mean_absolute_error
error = mean_absolute_error(test["medals"], test["predictions"])
error
```

```python
teams.describe()["medals"] #std deviation here must be higher than error calaculated above. A good check
```

```python
errors = (test["medals"]- test["predictions"]).abs()
errors
```

```python
error_by_team = errors.groupby(test["team"]).mean()
error_by_team
```

```python
medals_by_team = test["medals"].groupby(test["team"]).mean()
error_ratio = error_by_team/medals_by_team
```
